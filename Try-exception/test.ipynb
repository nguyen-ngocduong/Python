```python
import matplotlib.pyplot as plt
import numpy as np

# V s  hnh trn (mindmap) vi 4 chc nng ca TQNC
fig, ax = plt.subplots(figsize=(10, 10))
ax.axis("off")

# Thit lp tm v bn knh
center_x, center_y = 0.5, 0.5
radius = 0.3

# 4 chc nng chnh vi gc (tnh theo radian)
functions = [
    ("1. Lun gii", np.pi/2, "lightyellow"),      # Top (90)
    ("2. K tha", 0, "lightgreen"),               # Right (0)
    ("3. Minh chng", -np.pi/2, "lightsalmon"),    # Bottom (-90)
    ("4. nh v", np.pi, "lightpink")             # Left (180)
]

# V ng trn nn
circle = plt.Circle((center_x, center_y), radius, fill=False, color='lightgray', linewidth=2, linestyle='--')
ax.add_patch(circle)

# Tnh ton v tr cc chc nng trn ng trn
positions = []
for text, angle, color in functions:
    x = center_x + radius * np.cos(angle)
    y = center_y + radius * np.sin(angle)
    positions.append((x, y))
    
    # V cc  chc nng
    ax.text(x, y, text, ha="center", va="center", fontsize=11,
            fontweight="bold", 
            bbox=dict(facecolor=color, edgecolor="black", boxstyle="round,pad=0.5"))

# V mi tn ni 4 chc nng theo chiu kim ng h
for i in range(len(functions)):
    start_x, start_y = positions[i]
    end_x, end_y = positions[(i + 1) % len(functions)]
    
    # Tnh vector t start n end
    dx = end_x - start_x
    dy = end_y - start_y
    
    # iu chnh im bt u v kt thc  mi tn khng chng ln text
    offset = 0.08
    start_x += offset * dx / np.sqrt(dx**2 + dy**2)
    start_y += offset * dy / np.sqrt(dx**2 + dy**2)
    end_x -= offset * dx / np.sqrt(dx**2 + dy**2)
    end_y -= offset * dy / np.sqrt(dx**2 + dy**2)
    
    ax.annotate("",
                xy=(end_x, end_y),
                xytext=(start_x, start_y),
                arrowprops=dict(
                    arrowstyle="->",
                    color="darkblue",
                    lw=2.5,
                    connectionstyle="arc3,rad=0.2"
                ))

# Thm tiu   gia
ax.text(center_x, center_y, "TQNC\n4 Chc nng", ha="center", va="center", 
        fontsize=14, fontweight="bold", 
        bbox=dict(facecolor="white", edgecolor="darkblue", boxstyle="round,pad=0.5"))

# Thit lp gii hn v t l
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')

plt.title("S  hnh trn - 4 chc nng ca TQNC", fontsize=16, fontweight="bold", pad=20)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Cc bc trong vng trn
steps = [
    "Cu hi / Mc tiu\nnghin cu",
    "C s l thuyt /\nM hnh nghin cu",
    "Bi cnh &\nPhng php",
    "Kt qu &\nng gp",
    "Hn ch &\nHng nghin cu\n tip theo"
]

# V tr gc cho vng trn
angles = [0, 72, 144, 216, 288]  

# Ta  x, y cho tng bc
x = [10 * np.cos(a * 3.14 / 180) for a in angles]
y = [10 * np.sin(a * 3.14 / 180) for a in angles]

# V vng trn
fig, ax = plt.subplots(figsize=(7,7))

# V cc bc di dng node
for i in range(len(steps)):
    ax.scatter(x[i], y[i], s=1200, c="skyblue", edgecolors="black", zorder=3)
    ax.text(x[i], y[i], steps[i], ha="center", va="center", fontsize=10, weight="bold")

# V mi tn ni
for i in range(len(steps)):
    ax.annotate("",
                xy=(x[(i+1)%len(steps)], y[(i+1)%len(steps)]),
                xytext=(x[i], y[i]),
                arrowprops=dict(arrowstyle="->", lw=2))

# Ci t hin th
ax.set_aspect("equal")
ax.axis("off")
plt.title("Chu trnh tm tt mt cng trnh nghin cu", fontsize=14, weight="bold")
plt.show()
```
