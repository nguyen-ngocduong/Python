## Chuan hoa price

```python
df = data.copy()
print(df['price'][0].split('~'))
```

```python
def chuan_hoa_price(data):
    """
    chuan hoa du lieu price thanh 2 cot: total_price va price_per_square
    """
    total_price = []
    price_per_square = []
    for row in data['price']:
        row = row.split('~')
        for i in row:
            if 'tỷ' in i:
                total_price.append(i)
            else:
                price_per_square.append(i)
    return total_price, price_per_square

total_price, price_per_square = chuan_hoa_price(data)
print(total_price[0], price_per_square[0])
```

```python
df['total_price'] = total_price
df['price_per_square'] = price_per_square
df.head()
```

## Chuan hoa cot area

```python
df['area'][0].split(' Mặt tiền ')
```

```python
def chuan_hoa_area(data):
    """
    CHuan hoa cot area thanh 2 cot moi: Dien_tich va Mat_tien
    """
    dien_tich, mat_tien = [], []
    for row in data['area']:
        if ' Mặt tiền ' in row:
            row = row.split(' Mặt tiền ')
            dien_tich.append(row[0].strip())
            mat_tien.append(row[1].strip())
        else:
            dien_tich.append(row.strip())
            mat_tien.append(None)
    return dien_tich,mat_tien
dien_tich, mat_tien = chuan_hoa_area(df)
df['Dien_Tich'] = dien_tich
df['Mat_Tien'] = mat_tien
df.head()
```

```python
df.to_csv('batdongsan_data_chuan_hoa.csv', index= False)
```
