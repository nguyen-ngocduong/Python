```python
import pandas as pd
import numpy as np
import ast
```

```python
path_dmx = '/home/admin123/Desktop/dataocubuntu/documents/Python/Try-exception/DA/crawl_data/craw_data_tv_samsung/data/craw_dmx.csv'
path_hc = '/home/admin123/Desktop/dataocubuntu/documents/Python/Try-exception/DA/crawl_data/craw_data_tv_samsung/data/scrape_hc.csv'
path_meta = '/home/admin123/Desktop/dataocubuntu/documents/Python/Try-exception/DA/crawl_data/craw_data_tv_samsung/data/scrape_meta.csv'
path_pico = '/home/admin123/Desktop/dataocubuntu/documents/Python/Try-exception/DA/crawl_data/craw_data_tv_samsung/data/scrape_pico.csv'
data1 = pd.read_csv(path_dmx)
data2 = pd.read_csv(path_hc)
data3 = pd.read_csv(path_meta)
data4 = pd.read_csv(path_pico)
```

```python
#Chuan hoa cot price
def chuan_hoa_price(data):
    """
    dua data ve dang chuan
    """
    for i in range(len(data1)):
        data['price'][i] = data['price'][i][:-1]
    return data
data1 = chuan_hoa_price(data1)
data1.head()
```

```python
print((ast.literal_eval(data1['description'][0].replace("'", '"').replace('""', "'").replace('"', '"'))))
```

```python
def chuan_hoa_description(row):
    """
    Dua tung row ve dang dict
    """
    try:
        row = row.replace("'", '"').replace('""', "'").replace('"', '"')
        row = ast.literal_eval(row)
        return row
    except Exception as e:
        print(f'Loi o ham chuan_hoa_description: {e}')
data1['description'] = data1['description'].apply(chuan_hoa_description)
data1['description'].head()
```

```python
import json
# 5. Mở rộng cột dictionary thành các cột mới
# pd.json_normalize là cách hiệu quả để mở rộng dữ liệu JSON/dictionary
df_expanded = pd.json_normalize(data1['description'])

# 6. Kết hợp DataFrame ban đầu với các cột mới
df_final = pd.concat([data1.drop(columns='description'), df_expanded], axis=1)

# 7. Hiển thị kết quả
# In 5 dòng đầu tiên và tất cả các cột để kiểm tra
print(df_final.head())

# In tất cả các cột và thông tin cơ bản
print("\n--- DataFrame sau khi chuẩn hóa ---")
print(df_final.info())
```

```python
def chuan_hoa_scrape_meta(data3):
    def chuan_hoa_price(row):
        """
        dua data ve dang chuan
        """
        try:
            row = row.split()[0]
            return row
        except Exception as e:
            print(f'Loi o ham chuan hoa price: {e}')
    data3['price'] = data3['price'].apply(chuan_hoa_price)
    def chuan_hoa_rate(row):
        """
        chuan hoa tung dong rate
        """
        try:
            row = row.split()[0]
            return row
        except Exception as e:
            print(f'Loi o ham chuan_hoa_rate: {e}')
    data3['rate'] = data3['rate'].apply(chuan_hoa_rate)
    def chuan_hoa_description_hc(row):
        """
        Chuan hoa tung hang cho description
        """
        try:
            row = row.replace("'", '"').replace('""', "'").replace('"', '"').replace("'", '"')
            row = ast.literal_eval(row)
            return row
        except Exception as e:
            print(f'Loi o ham chuan_hoa_description_hc: {e}')
    data3['description'] = data3['description'].apply(chuan_hoa_description_hc)
    # pd.json_normalize là cách hiệu quả để mở rộng dữ liệu JSON/dictionary
    df_expanded = pd.json_normalize(data3['description'])

    # 6. Kết hợp DataFrame ban đầu với các cột mới
    df_final = pd.concat([data3.drop(columns='description'), df_expanded], axis=1)

    # 7. save ket qua
    df_final.to_csv('/home/admin123/Desktop/dataocubuntu/documents/Python/Try-exception/DA/crawl_data/craw_data_tv_samsung/data/scrape_meta_cleaned.csv', index=False)
```
